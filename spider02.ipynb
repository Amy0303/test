# 添加headers,header为请求头部信息
 - http://www.zhihu.com/explore
 - 最基础也需要填写一个User-Agent

```python
import requests
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/66.0.3359.181 Safari/537.36'}
response = requests.get('https://www.zhihu.com/explore',headers=header)
print(response.text)
```

# response属性
- allow_redirects = False不允许跳转
- status_code**
- headers**
- cookies**
- url**
- history**
- 'http://www.jianshu.com'

```python
import requests
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/66.0.3359.181 Safari/537.36'}
response = requests.get('https://www.jianshu.com',headers=header,allow_redirects=True)
print(response.text)
print(response.url)
print(response.histroy)
```

```python
def main():
    import requests
    a = ['http://zhuanti.17k.com/subject/2018/newbooks/juewangliming/pc/','http://zhuanti.17k.com/subject/2018/wusong/pc/']
    header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/66.0.3359.181 Safari/537.36'}
    for i in range(2):
        url = a[i]
        response = requests.get(url,headers=header)
        file = 'C:\\Users\\Administrator\\Desktop\\e'+str(i)+'.html'
        with open(file,'w',encoding='utf-8')as f:
            f.write(response.text)
        print(i)
```
